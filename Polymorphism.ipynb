```python
class Employee:
    def __init__(self, name, city, ENum):
        self.name = name 
        self.city = city
        self.ENum = ENum
        
    def __str__(self):
        s = f"Name: {self.name}\n"
        s += f"City: {self.city}\n"
        s += f"Employee Number: {self.ENum}\n"
        return s
    
    def title(self, job):
        return f"{self.name} works as a {job}"
    
    def manager(self):
        if self.ENum % 2 == 0:
            return f"{manager1.name} is {self.name}'s manager"
        else:
            return f"{manager2.name} is {self.name}'s manager"
    
class Manager(Employee):
    def __init__(self, name, city, MNum):
        super().__init__(name, city, MNum)
        self.MNum = MNum 
  
    def __str__(self):
        s = f"Name: {self.name}\n"
        s += f"City: {self.city}\n"
        s += f"Manager Number: {self.MNum}\n"
        return s
    
    def manager(self):
        if self.MNum % 2 == 0:
            return f"{manager1.name} is {employee1.name}'s manager"
        else:
            return f"{manager2.name} is {employee2.name}'s manager"
        
# Employee and manager instances:
employee1 = Employee("Orlaigh King", "Dublin", 1)
employee2 = Employee("Anna White", "Kildare", 2)
manager1 = Manager("Dave O'Brien", "Belfast", 3)
manager2 = Manager("Lola Kenny", "Kildare", 4)
print()

# Print information and manager details:
print(employee1)
print(employee2)
print(employee1.title("IT specialist"))
print(manager2.title("Head of Finance"))
print()
print(employee1.manager())
print(employee2.manager())
print()
print(manager1)
print(manager2)
print(manager1.manager())
print(manager2.manager())

```
